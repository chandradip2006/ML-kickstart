```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Y23 UG SPI_CPI IITK.csv")
df
```

```python
df.info()
```

```python
df=df.drop(['Roll Number' , 'Name' , 'Department' , '2023-24 3' , '2024-25 2'] , axis=1)
df
```

```python
spi=df.dropna(axis=0)
spi.info()
```

```python
%matplotlib inline
plt.scatter(spi['2023-24 1'] , spi['2024-25 1'] , color='red' , marker='.')
```

```python
%matplotlib inline
plt.scatter(spi['2023-24 2'] , spi['2024-25 1'] , color='red' , marker='.')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(spi[['2023-24 1' , '2023-24 2']] , spi['2024-25 1'])
```

```python
reg.score(spi[['2023-24 1' , '2023-24 2']] , spi['2024-25 1'])
```

```python
reg.predict([[9.3 , 9.5]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
