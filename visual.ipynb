```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def gradient_decent(x , y):
    m_curr=b_curr=0
    n=len(x)
    itr=10000
    learning=0.08
    y_predicted=np.arange(5)

    for i in range(itr):
        y_predicted=m_curr*x+b_curr
        cost=(1/n)*sum((y-y_predicted)**2)
        print(f"m:{m_curr} b:{b_curr} cost:{cost} itr:{i}")
        m_dec=(-1)*(2/n)*sum(x*(y-y_predicted))
        b_dec=(-1)*(2/n)*sum(y-y_predicted)

        m_curr=m_curr-learning*m_dec
        b_curr=b_curr-learning*b_dec

        # %matplotlib inline
        # plt.plot(x , y_predicted)
    return y_predicted
```

```python
x=np.array([1 , 2 , 3 , 4 , 5])
y=np.array([5 , 7 , 9 , 11 , 13])
# y_pre=np.arange(5)

y_pre=gradient_decent(x , y)
# y_pre=np.array(y_pre)

%matplotlib inline
plt.scatter(x , y , color='red' , marker='+')
plt.plot(x , y_pre , color='blue')
```
