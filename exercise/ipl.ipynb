```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ipl.csv")
df
```

```python
df.info()
```

```python
df.runs.skew()
```

```python
dummies1=pd.get_dummies(df.venue)
# dummies
dummies1=dummies1.astype(int)
dummies1
```

```python
dummies1.columns
```

```python
dummies2=pd.get_dummies(df.bat_team)
# dummies
dummies2=dummies2.astype(int)
dummies2
```

```python
dummies2.columns
```

```python
dummies3=pd.get_dummies(df.bowl_team)
# dummies
dummies3=dummies3.astype(int)
dummies3
```

```python
dummies3.columns
```

```python
last=pd.concat([df , dummies1 , dummies2 , dummies3] , axis='columns')
last
```

```python
last.info()
```

```python
last.columns
```

```python
last.info()
```

```python
total=last['total']
total
```

```python
merge=last.drop(['mid' , 'date' , 'venue' , 'bat_team' , 'bowl_team' , 'batsman' , 'bowler' , 'runs' , 'wickets' , 'overs' , 'runs_last_5' , 'wickets_last_5' , 'striker' , 'non-striker' , 'total'] , axis='columns')
merge
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(merge , total)
```

```python
reg.score(merge , total)
```
